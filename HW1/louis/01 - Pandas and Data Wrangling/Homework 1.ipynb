## Task 1. Compiling Ebola Data

The `DATA_FOLDER/ebola` folder contains summarized reports of Ebola cases from three countries (Guinea, Liberia and Sierra Leone) during the recent outbreak of the disease in West Africa. For each country, there are daily reports that contain various information about the outbreak in several cities in each country.

Use pandas to import these data files into a single `Dataframe`.
Using this `DataFrame`, calculate for *each country*, the *daily average per month* of *new cases* and *deaths*.
Make sure you handle all the different expressions for *new cases* and *deaths* that are used in the reports.

```python
import os
guinea_list=[]
pathname = os.path.join(DATA_FOLDER+"\\ebola", 'guinea_data')
for csv_file in os.listdir(pathname):
    abs_path=os.path.join(pathname,csv_file)
    df=pd.read_csv(abs_path,index_col=0,parse_dates=[0])
    guinea_list.append(df)
df=pd.concat(guinea_list)
df['Month']=df.index.month
df
```

```python
df_reduced=df[['Month','Description','Totals']]
interesting_metrics_description=['New cases','New deaths']

# Change "new deaths registered today" of first day to "new deaths". same for cases
new_description=df_reduced['Description']
new_description[new_description=='New deaths registered today']='New deaths'
new_description[new_description=='New deaths registered']='New deaths'
new_description[new_description=='Total new cases registered so far']='New cases'
df_reduced['Description']=new_description

# Getting number of days in month
df_reduced['Date']=df_reduced.index
df_reduced=df_reduced.set_index(['Description'])
months=df_reduced.loc[interesting_metrics_description[0]]
months=months['Month']
months=months.value_counts()
months=pd.DataFrame(months)
months=months.rename(columns={'Month':'Days_per_month'})
months['Month']=months.index
print(months)

# Getting relevant descriptions
df_reduced_only_cool_metrics=df_reduced.loc[interesting_metrics_description]

df_reduced_only_cool_metrics=df_reduced_only_cool_metrics.reset_index().set_index(['Description','Month','Date'])
df_reduced_only_cool_metrics['Totals']=pd.to_numeric(df_reduced_only_cool_metrics['Totals'])

df_summed=df_reduced_only_cool_metrics.sum(level=['Description','Month'])

df_summed=df_summed.reset_index()
df_merged=pd.merge(df_summed,months,on='Month',how='left')

df_merged=df_merged.set_index(['Description','Month'])
df_merged['Totals']=df_merged['Totals']/df_merged['Days_per_month']
df_divided=df_merged.drop('Days_per_month',axis=1)
df_guinea=df_divided
df_guinea
# Check number of days
```

```python
import os
liberia_list=[]
pathname = os.path.join(DATA_FOLDER+"\\ebola", 'liberia_data')
for csv_file in os.listdir(pathname):
    abs_path=os.path.join(pathname,csv_file)
    df=pd.read_csv(abs_path,index_col=0,parse_dates=[0])
    liberia_list.append(df)
df=pd.concat(liberia_list)
df['Month']=df.index.month
```

```python
df_reduced=df[['Month','Variable','National']]
df_reduced['National']=pd.to_numeric(df_reduced['National'])
print(df_reduced.shape)
print(df_reduced.dropna().shape)
# Are we sure ???
df_reduced=df_reduced.dropna()

# Newly reported deaths
# New Case/s (Suspected)
# New Case/s (Probable)
# New case/s (confirmed)

df_reduced['Date']=df_reduced.index
df_reduced_descr=df_reduced.set_index('Variable')

months=df_reduced_descr.loc['Newly reported deaths']
months=months['Month']
months=months.value_counts()
months=pd.DataFrame(months)
months=months.rename(columns={'Month':'Days_per_month'})
months['Month']=months.index

df_reduced_selected_cases=df_reduced_descr.loc[['New Case/s (Suspected)','New Case/s (Probable)','New Case/s (confirmed)']]
df_reduced_selected_cases=df_reduced_selected_cases.reset_index().set_index(['Month','Variable','Date'])
df_summed_cases=df_reduced_selected_cases.sum(level=[0])
df_summed_cases['Variable']=['New Cases']*df_summed_cases.shape[0]
df_summed_cases=df_summed_cases.reset_index()

df_deaths=df_reduced_descr.loc['Newly reported deaths']
df_deaths=df_deaths.reset_index().set_index(['Month','Variable','Date'])
df_summed_death=df_deaths.sum(level=[0,1])
df_summed_death=df_summed_death.reset_index()
df_summed_death['Variable']=['New deaths']*df_summed_death.shape[0]

df_summed=pd.concat([df_summed_cases,df_summed_death])

df_merged=pd.merge(df_summed,months,on='Month',how='left')
df_merged=df_merged.rename(columns={'Variable':'Description'})
df_merged=df_merged.set_index(['Description','Month'])
df_merged['Totals']=df_merged['National']/df_merged['Days_per_month']
df_divided=df_merged.drop(['Days_per_month','National'],axis=1)
df_divided=df_divided.dropna()
df_liberia=df_divided
df_liberia
```

```python
import os
sl_list=[]
pathname = os.path.join(DATA_FOLDER+"\\ebola", 'sl_data')
for csv_file in os.listdir(pathname):
    abs_path=os.path.join(pathname,csv_file)
    df=pd.read_csv(abs_path,index_col=0,parse_dates=[0])
    sl_list.append(df)
df=pd.concat(sl_list)
df['Month']=df.index.month
df
```

```python
df_reduced=df[['Month','variable','National']]
df_reduced['National']=pd.to_numeric(df_reduced['National'],errors='coerce')

df_reduced['Date']=df_reduced.index
df_reduced_descr=df_reduced.set_index('variable')

months=df_reduced_descr.loc['death_suspected']
months=months['Month']
months=months.value_counts()
months=pd.DataFrame(months)
months=months.rename(columns={'Month':'Days_per_month'})
months['Month']=months.index

shifter1=df_reduced_descr.loc['death_suspected'] 
shifter1=shifter1.reset_index()
tmp=list(shifter1['National'])
tmp[1:]=tmp[:-1]
shifter1=pd.concat([shifter1,pd.DataFrame(list(tmp),columns=['to_remove'])],axis=1)
shifter1['National']=shifter1['National']-shifter1['to_remove']
shifter1=shifter1.drop('to_remove',axis=1)

shifter2=df_reduced_descr.loc['death_probable']  
shifter2=shifter2.reset_index()
tmp=list(shifter2['National'])
tmp[1:]=tmp[:-1]
shifter2=pd.concat([shifter2,pd.DataFrame(list(tmp),columns=['to_remove'])],axis=1)
shifter2['National']=shifter2['National']-shifter2['to_remove']
shifter2=shifter2.drop('to_remove',axis=1)

shifter3=df_reduced_descr.loc['death_confirmed']  
shifter3=shifter3.reset_index()
tmp=list(shifter3['National'])
tmp[1:]=tmp[:-1]
shifter3=pd.concat([shifter3,pd.DataFrame(list(tmp),columns=['to_remove'])],axis=1)
shifter3['National']=shifter3['National']-shifter3['to_remove']
shifter3=shifter3.drop('to_remove',axis=1)


df_shifted=pd.concat([shifter1,shifter2,shifter3])


df_shifted=df_shifted.set_index('Month')
df_deaths=df_shifted.sum(level='Month')
df_deaths['Description']=['New deaths']*df_deaths.shape[0]
df_deaths=df_deaths.reset_index()


df_cases=df_reduced_descr.loc[['new_suspected','new_probable','new_confirmed']]
df_cases=df_cases.reset_index().set_index('Month')
df_cases=df_cases.sum(level='Month')
df_cases['Description']=['New cases']*df_cases.shape[0]
df_cases=df_cases.reset_index()

df_selected=pd.concat([df_cases,df_deaths])
df_selected=df_selected.set_index(['Description','Month'])
df_summed=df_selected.sum(level=[0,1])
df_summed=df_summed.reset_index()

df_merged=pd.merge(df_summed,months,on='Month',how='left')
df_merged=df_merged.set_index(['Description','Month'])
df_merged['Totals']=df_merged['National']/df_merged['Days_per_month']
df_divided=df_merged.drop(['Days_per_month','National'],axis=1)
df_divided=df_divided.dropna()
df_sl=df_divided
df_sl
```

```python
df_guinea['Country']=['Guinea']*df_guinea.shape[0]
df_liberia['Country']=['Liberia']*df_liberia.shape[0]
df_sl['Country']=['SL']*df_sl.shape[0]

df_full=pd.concat([df_guinea.reset_index(),df_liberia.reset_index(),df_sl.reset_index()])
df_full=df_full.set_index(['Country','Description','Month'])
df_full=df_full.rename(columns={'Totals':"Daily average per Month"})
df_full
```

 ### 1. Describe the *type* and the *value range* of each attribute. Indicate and transform the attributes that can be `Categorical`.

```python
# Show the classes of all attributes
df_titanic.dtypes
```

```python
import numpy as np

# For all attributes list the different values
for attribute in df_titanic.columns:
    unique_values = df_titanic[attribute].unique()
    print('All unique values for %s are : %s' % (attribute, unique_values))
    
    # Only show the range when it is numerical values
    if ((unique_values.dtype == np.int64) or (unique_values.dtype == np.float64)):
        unique_values = unique_values[np.logical_not(np.isnan(unique_values))]
        print('Values range from %s to %s' % (unique_values.min(), unique_values.max()))
    print('\n\n')
```

All attributes that can take a finite number of values are caterogical, i.e. 'pclass', 'survived', 'sex', 'sibsp', 'parch', 'embarked' and 'boat'. We will transform them to categorical type.

```python
cat_col = ['pclass', 'survived', 'sex', 'sibsp', 'parch', 'embarked', 'boat']
for col in cat_col:
    df_titanic[col] = df_titanic[col].astype('category')
print(df_titanic.dtypes)
```

### 2. Plot histograms for the *travel class*, *embarkation port*, *sex* and *age* attributes. For the latter one, use *discrete decade intervals*. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

hist_attributes = ['pclass', 'embarked', 'sex']
for attr in hist_attributes:
    plt.figure();
    sns.countplot(df_titanic[attr])
    plt.show()
```

### 4. For each *travel class*, calculate the proportion of the passengers that survived. Present your results in *pie charts*.

```python
# Create an array with all the classes
classes = df_titanic.pclass.unique()
for tr_class in classes:
    print('Class %s' % tr_class)
    cl_survived = df_titanic[df_titanic['pclass'] == tr_class].survived
    cl_survived.value_counts().plot.pie(autopct='%.2f', fontsize=20, figsize=(6, 6))
    plt.show()
    print('\n\n')
```
