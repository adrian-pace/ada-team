## Task 1. Compiling Ebola Data

The `DATA_FOLDER/ebola` folder contains summarized reports of Ebola cases from three countries (Guinea, Liberia and Sierra Leone) during the recent outbreak of the disease in West Africa. For each country, there are daily reports that contain various information about the outbreak in several cities in each country.

Use pandas to import these data files into a single `Dataframe`.
Using this `DataFrame`, calculate for *each country*, the *daily average per month* of *new cases* and *deaths*.
Make sure you handle all the different expressions for *new cases* and *deaths* that are used in the reports.

```python
import os
# Fetching the files
all_dataframes=[]

for country_dir in os.listdir(DATA_FOLDER+"\\ebola"):
    pathname = os.path.join(DATA_FOLDER+"\\ebola", country_dir)
    # We look at each country folder
    if os.path.isdir(pathname):
        # we recover all the dataframe for a particular country
        for csv_file in os.listdir(pathname):
            abs_path=os.path.join(pathname,csv_file)
            df=pd.read_csv(abs_path,index_col=0,parse_dates=[0])
            df['Country']=[country_dir]*df.shape[0]
            all_dataframes.append(df)
```

```python
all_dataframes[0]
```

```python
df=pd.concat(all_dataframes)
# Get the month
df['Month']=df.index.month
# Concatenate all the dfs
```
